This research employed a binary variable, default payment (Yes = 1, No = 0), as the response variable. This study reviewed the literature and used the following 23 variables as explanatory variables:

X1: Amount of the given credit (NT dollar): it includes both the individual consumer credit and his/her family (supplementary) credit.

X2: Gender (1 = male; 2 = female).

X3: Education (1 = graduate school; 2 = university; 3 = high school; 4 = others).

X4: Marital status (1 = married; 2 = single; 3 = others).

X5: Age (year).

X6 - X11: History of past payment. We tracked the past monthly payment records (from April to September, 2005) as follows: X6 = the repayment status in September, 2005;

X7 = the repayment status in August, 2005; . . .;X11 = the repayment status in April, 2005. The measurement scale for the repayment status is: -1 = pay duly; 1 = payment delay for one month; 2 = payment delay for two months; . . .; 8 = payment delay for eight months; 9 = payment delay for nine months and above.

X12-X17: Amount of bill statement (NT dollar). X12 = amount of bill statement in September, 2005; X13 = amount of bill statement in August, 2005; . . .; X17 = amount of bill statement in April, 2005.

X18-X23: Amount of previous payment (NT dollar). X18 = amount paid in September, 2005; X19 = amount paid in August, 2005; . . .;X23 = amount paid in April, 2005.

# Normalizing data

```python
# # import Scalers
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Define feature groups
nt_dollar_features = ["X1"] + [f"X{i}" for i in range(12, 24)]  # X1, X12-X23 (Monetary values)
repayment_status_features = [f"X{i}" for i in range(6, 12)]     # X6-X11 (Repayment history)

# Initialize scalers
scaler_standard = StandardScaler()  # For NT dollar values
scaler_minmax = MinMaxScaler()      # For Repayment history

# Apply transformations
X[nt_dollar_features] = scaler_standard.fit_transform(X[nt_dollar_features])
X[repayment_status_features] = scaler_minmax.fit_transform(X[repayment_status_features])

# Results
print(X.head())
```

```python
from sklearn.model_selection import train_test_split

# Split your dataset into 80/20 dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size= 0.2)

# Split your dataset into 60/40 dataset
#X_train, X_test, y_train, y_test = train_test_split(X,y,test_size= 0.4)

# Split your dataset into 70/30 dataset
#X_train, X_test, y_train, y_test = train_test_split(X,y,test_size= 0.3)

# Verify shape of test/train sets
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Validating KNN Classifier

```python
# Import libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
import numpy as np

# Define the list of n_neighbors values to test
knn_neighbors = [3, 4, 5, 7, 10, 20, 50]

# Dictionary to store results
knn_results = {}

# Iterate over each value of k and perform cross-validation
for k in knn_neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=4, scoring='precision')

    # Store the mean precision score
    knn_results[k] = np.mean(scores)

    # Print individual cross-validation scores
    print(f"k = {k}, Cross-validation scores: {scores}, Mean Precision: {np.mean(scores):.4f}")

# Find the best k based on the highest precision score
best_k = max(knn_results, key=knn_results.get)
best_precision = knn_results[best_k]

print("\nKNN Precision Scores Summary:", knn_results)
print(f"Best k: {best_k} with Precision: {best_precision:.4f}")

```

## Generating ROC curve for KNN classifier

Using K=50 due to the best precision score

```python
from sklearn.metrics import roc_curve, auc

# Using best k from cross-validation
best_knn = KNeighborsClassifier(n_neighbors=best_k)

# Train the model on the full training set
best_knn.fit(X_train, y_train)

# Get predicted probabilities for the positive class (class 1)
# ?????????? NEED to RE-CHECK to better understand (Ievgen)
y_prob_knn = best_knn.predict_proba(X_test)[:, 1]

# Compute ROC curve
fpr_knn, tpr_knn, thresholds_knn = roc_curve(y_test, y_prob_knn)

# Compute AUC (Area Under the Curve)
roc_auc_knn = auc(fpr_knn, tpr_knn)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr_knn, tpr_knn, color='blue', lw=2, label=f'ROC curve (AUC = {roc_auc_knn:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')  # Random classifier line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve for KNN classifier')
plt.legend(loc='lower right')
plt.show()
```

```python
# Print the thresholds
print("Thresholds:", thresholds_knn)
```

```python
# Print the thresholds with their corresponding FPR and TPR
for t, f, tp in zip(thresholds_knn, fpr_knn, tpr_knn):
    print(f"Threshold: {t:.4f} | FPR: {f:.4f} | TPR: {tp:.4f}")
```

## Finding best threshold for KNN Classifier

```python
# Calculate Youden's J statistic for each threshold
j_statistic = tpr_knn - fpr_knn

# Find the index of the maximum J statistic
best_threshold_index_knn = np.argmax(j_statistic)

# Get the best threshold
best_threshold_knn = thresholds_knn[best_threshold_index_knn]

print(f"Optimal threshold: {best_threshold_knn:.4f}")
```

## Applying Threshold and creating confusion matrix for KNN Classifier

```python
y_pred_knn = (y_prob_knn >= best_threshold_knn).astype(int)
```

```python
from sklearn.metrics import confusion_matrix

# Confusion Matrix
cm_knn = confusion_matrix(y_test, y_pred_knn)
print("Confusion Matrix:")
print(cm_knn)
```

# Validating Random Forest Classifier

```python
# Import libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np

# Define the range of n_estimators to tune
n_estimators_values = [3, 4, 6, 7, 10, 20, 50, 100]

# Dictionary to store results
rf_results = {}

# Loop through the different n_estimators values
for n in n_estimators_values:
    # Initialize the Random Forest Classifier with the current value of n_estimators
    rf_model = RandomForestClassifier(n_estimators=n, random_state=123)

    # Perform 4-fold cross-validation using precision as the scoring metric
    scores_rf = cross_val_score(rf_model, X_train, y_train, cv=4, scoring='precision')

    # Store the mean precision score
    rf_results[n] = np.mean(scores_rf)

    # Print individual cross-validation scores
    print(f"n = {n}, Cross-validation scores: {scores_rf}, Mean Precision: {np.mean(scores_rf):.4f}")

# Find the best n based on the highest precision score
best_n_rf = max(rf_results, key=rf_results.get)
best_precision_rf = rf_results[best_n_rf]

print("\nRF Precision Scores Summary:", rf_results)
print(f"Best n: {best_n_rf} with Precision: {best_precision_rf:.4f}")
```

## Generating ROC curve for RF classifier

```python
from sklearn.metrics import roc_curve, auc

# Using best n from cross-validation
best_rf = RandomForestClassifier(n_estimators=best_n_rf)

# Train the model on the full training set
best_rf.fit(X_train, y_train)

# Get predicted probabilities for the positive class (class 1)
# ?????????? NEED to RE-CHECK to better understand (Ievgen)
y_prob_rf = best_rf.predict_proba(X_test)[:, 1]

# Compute ROC curve
fpr_rf, tpr_rf, thresholds_rf = roc_curve(y_test, y_prob_rf)

# Compute AUC (Area Under the Curve)
roc_auc_rf = auc(fpr_rf, tpr_rf)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr_rf, tpr_rf, color='blue', lw=2, label=f'ROC curve (AUC = {roc_auc_rf:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')  # Random classifier line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve for RF classifier')
plt.legend(loc='lower right')
plt.show()
```

## Finding best threshold for RF Classifier

```python
# Calculate Youden's J statistic for each threshold
j_statistic_rf = tpr_rf - fpr_rf

# Find the index of the maximum J statistic
best_threshold_index_rf = np.argmax(j_statistic_rf)

# Get the best threshold
best_threshold_rf = thresholds_rf[best_threshold_index_rf]

print(f"Optimal threshold for RF: {best_threshold_rf:.4f}")
```

## Applying Threshold and creating confusion matrix for RF Classifier

```python
y_pred_rf = (y_prob_rf >= best_threshold_rf).astype(int)
```

```python
from sklearn.metrics import confusion_matrix

# Confusion Matrix
cm_rf = confusion_matrix(y_test, y_pred_rf)
print("Confusion Matrix:")
print(cm_rf)
```
